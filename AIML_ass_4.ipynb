```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
housing = pd.read_csv('/content/Housing.csv')
print (housing.head())
housing.shape
housing.info()
housing.describe()
```

```python
print(housing.dtypes)
```

```python
housing.describe(percentiles=[.25, .50, .75, .90,])
```

```python
print(housing.columns)
```

```python
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus'] # Add 'furnishingstatus' to the list

def binary_map(x):
    return x.map({'yes': 1, "no": 0, 'furnished':1, 'semi-furnished':0.5, 'unfurnished':0}) # Add mappings for values in 'furnishingstatus'

housing[varlist] = housing[varlist].apply(binary_map)
```

```python
housing.head()
```

```python
import seaborn as sns
sns.pairplot(housing)
```

```python
housing.corr()
```

```python
import matplotlib.pyplot as plt

corr_matrix = housing.corr()
print(corr_matrix)


plt.figure(figsize=(16, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
